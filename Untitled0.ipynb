```python
import torch
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image
from tqdm import tqdm
import numpy as np
import os

# === Paths ===
dataset_path = "/content/dataset/lfw_250_classes"
embedding_file = "/content/embeddings.npz"

# === Models ===
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
mtcnn = MTCNN(image_size=160, margin=10, min_face_size=20, device=device)
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)

print(f"✅ Using device: {device}")

# === Prepare lists ===
embeddings = []
labels = []

# === Loop through each person folder ===
for person in tqdm(os.listdir(dataset_path), desc="Processing people"):
    person_path = os.path.join(dataset_path, person)
    if not os.path.isdir(person_path):
        continue

    for img_name in os.listdir(person_path):
        if not img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        img_path = os.path.join(person_path, img_name)
        try:
            img = Image.open(img_path).convert('RGB')
            face = mtcnn(img)
            if face is not None:
                embedding = resnet(face.unsqueeze(0).to(device)).detach().cpu().numpy()
                embeddings.append(embedding)
                labels.append(person)
        except Exception as e:
            print(f"⚠️ Error with {img_path}: {e}")

# === Save embeddings ===
embeddings = np.concatenate(embeddings)
np.savez(embedding_file, embeddings=embeddings, labels=np.array(labels))
print(f"\n✅ Saved embeddings to: {embedding_file}")
print(f"Total faces embedded: {len(labels)}")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Path to dataset
data_dir = "/content/dataset/lfw_250_classes/LFW_Peopels"

# Image size and batch size
IMG_SIZE = (128, 128)
BATCH_SIZE = 32

# Split and preprocess
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_gen = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    data_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

print("✅ Train and validation generators ready.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

num_classes = len(train_gen.class_indices)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.4),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam

# Load MobileNetV2 with pretrained ImageNet weights
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(128,128,3))

# Freeze base layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

# Define final model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile
model.compile(optimizer=Adam(learning_rate=0.0005), loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# ✅ Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    validation_data=val_generator,
    validation_steps=len(val_generator),
    epochs=15,
    verbose=1
)
```

```python
import os, numpy as np, torch
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image
from tqdm import tqdm
import joblib

# Paths
data_root = "/content/dataset/lfw_250_classes/LFW_Peopels"
emb_file = "/content/face_embeddings.npz"

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:", device)

# init detector + embedder
mtcnn = MTCNN(image_size=160, margin=10, keep_all=False, device=device)
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)

labels = []
embs = []

# iterate classes
classes = sorted([d for d in os.listdir(data_root) if os.path.isdir(os.path.join(data_root,d))])
print("Found classes:", len(classes))

for cls in tqdm(classes, desc="Classes"):
    cls_path = os.path.join(data_root, cls)
    imgs = sorted([f for f in os.listdir(cls_path) if f.lower().endswith(('.jpg','.jpeg','.png'))])
    for fn in imgs:
        p = os.path.join(cls_path, fn)
        try:
            img = Image.open(p).convert('RGB')
        except:
            continue
        # detect & extract face tensor
        face_tensor = mtcnn(img)  # returns tensor (3,160,160) or None
        if face_tensor is None:
            continue
        with torch.no_grad():
            emb = resnet(face_tensor.unsqueeze(0).to(device)).cpu().numpy().reshape(-1)
        embs.append(emb)
        labels.append(cls)

embs = np.array(embs)
labels = np.array(labels)
print("Embeddings shape:", embs.shape, "Labels:", labels.shape)

np.savez(emb_file, embeddings=embs, labels=labels)
print("Saved embeddings to:", emb_file)
```

```python
import numpy as np
import os
from tqdm import tqdm
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# ✅ Path to your dataset
base_dir = "/content/dataset/lfw_250_classes/LFW_Peopels"

# ✅ Load lightweight CNN (MobileNetV2)
model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')

embeddings, labels = [], []
print("🔍 Extracting embeddings (this may take a few minutes)...")

for person in tqdm(os.listdir(base_dir)):
    person_dir = os.path.join(base_dir, person)
    if not os.path.isdir(person_dir):
        continue
    for img_name in os.listdir(person_dir):
        img_path = os.path.join(person_dir, img_name)
        try:
            img = image.load_img(img_path, target_size=(160, 160))
            x = image.img_to_array(img)
            x = np.expand_dims(x, axis=0)
            x = preprocess_input(x)
            feat = model.predict(x, verbose=0)
            embeddings.append(feat.flatten())
            labels.append(person)
        except:
            continue

X = np.array(embeddings)
y = np.array(labels)
print(f"✅ Extracted {X.shape[0]} embeddings with shape {X.shape[1:]}")

# Encode + split
le = LabelEncoder()
y_enc = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=0.2, stratify=y_enc, random_state=42)

# Train simple SVM
print("\n🚀 Training SVM classifier...")
clf = SVC(kernel='linear', probability=True)
clf.fit(X_train, y_train)

# Evaluate
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\n🎯 Accuracy: {acc*100:.2f}%")

# Save
joblib.dump(clf, "/content/face_classifier.joblib")
np.savez("/content/lfw_embeddings_final.npz", embeddings=X, labels=y)
print("💾 Model and embeddings saved successfully.")
```
